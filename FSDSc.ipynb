###**Pyplot vs. OO plotting**
 Here's an example that demonstrates the different plotting approaches in Python using both the pyplot (state-based) and the object-oriented (OO) interfaces.

**Using Pyplot (State-based):**

import matplotlib.pyplot as plt

#### **Data**
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

####**Create a line plot using pyplot**
**plt.plot(x, y, 'r--')**

> Indented block

**plt.xlabel('X-axis')**
**plt.ylabel('Y-axis')**
**plt.title('Pyplot Line Plot')**
**plt.show()**

In this example, we import the pyplot module from the matplotlib library and use the various pyplot functions to create a line plot. The plot() function is used to plot the data, and xlabel(), ylabel(), and title() functions are used to set the axis labels and the plot title.
**Note**
The **'r--' argument in the plot()** function represents the line style and color of the plotted line.

In this case, 'r--' is a shorthand notation where 'r' stands for red color, and '--' stands for a dashed line style.

Here's a breakdown of the shorthand notation:

**'r' represents the color red**.
'--' represents a dashed line style. **bold text**

You can use different characters and combinations to specify different line styles and colors. For example:

'b-' would represent a blue solid line.

'g:' would represent a green dotted line.
'k-.' would represent a black dash-dot line.

You can refer to the Matplotlib documentation for more options and customization possibilities for line styles and colors.

**Using Object-Oriented (OO) Approach:**

import matplotlib.pyplot as plt

#### **Data**
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

#### **Create a figure and axes objects**
**fig, ax = plt.subplots()**

#### Plot the data using the axes object
**ax.plot(x, y, 'b-')**
**ax.set_xlabel('X-axis')**
**ax.set_ylabel('Y-axis')**
**ax.set_title('OO Line Plot')**

#### **Display the plot**
**plt.show()**

In the OO approach, we create a figure and an axes object using the subplots() function.
Then, we use the plot() method of the axes object to plot the data. The set_xlabel(), set_ylabel(), and set_title() methods are used to set the labels and title.

Finally, we display the plot using plt.show().

Both approaches achieve the same result, but the OO approach gives you more flexibility and control over the plot elements, making it suitable for complex or multi-panel plots. The pyplot approach, on the other hand, provides a more concise syntax and is useful for quick and simple plots.

###Converting the Price column in the car_sales dataframe to an integer
Let's break down the line of code you mentioned:

**car_sales['Price'] = car_sales['Price'].str.replace('[^\d]', '',  regex=True).astype(int)**

**car_sales['Price'] **
This selects the "Price" column in the car_sales DataFrame, allowing us to operate specifically on that column.

**.str.replace('[^\d]', '', regex=True)**
This method call applies a regular expression to each value in the "Price" column. The regular expression [^\d] matches any non-digit character. By specifying regex=True, we enable regular expression matching. The matched non-digit characters are replaced with an empty string, effectively removing them from the values.

**.astype(int) **
This converts the values in the "Price" column to integers. After removing the non-digit characters, the remaining string values can be safely cast to integers.

Putting it all together, this line of code removes any non-digit characters from the "Price" column using regular expressions, and then converts the cleaned values to integers using .astype(int). The resulting column will contain integer values that can be used for mathematical calculations.


---
**Note**

**A regular expression, often referred to as regex or regexp, is a sequence of characters that forms a search pattern. It is used to match and manipulate text strings based on specific patterns.** Regular expressions are widely used in programming, text processing, and data validation tasks.

In regular expressions, certain characters and combinations have special meanings. Here are some commonly used characters and their meanings:

\d: Matches any digit (0-9).
\D: Matches any non-digit character.
\w: Matches any alphanumeric character (a-z, A-Z, 0-9, and underscore).
\W: Matches any non-alphanumeric character.
\s: Matches any whitespace character (spaces, tabs, line breaks).
\S: Matches any non-whitespace character.
.: Matches any character except a newline.
[]: Matches any single character within the brackets.
^: Matches the start of a string.
$: Matches the end of a string.
*: Matches zero or more occurrences of the preceding character or group.
+: Matches one or more occurrences of the preceding character or group.
?: Matches zero or one occurrence of the preceding character or group.
{n}: Matches exactly n occurrences of the preceding character or group.
{n,}: Matches n or more occurrences of the preceding character or group.
{n,m}: Matches between n and m occurrences of the preceding character or group.
These are just a few examples of the characters and syntax used in regular expressions. Regular expressions can become complex and powerful tools for pattern matching and text manipulation.

**In the context of the code snippet we discussed earlier, the regular expression [^\d] is used to match any non-digit character in the "Price" column. The str.replace() method replaces those matched characters with an empty string, effectively removing them from the values.**

###Notes on the OO graph method

The given code snippet is related to creating a scatter plot using the Matplotlib library. Let's break down the snippet:

**scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])**

**ax0:**
It represents an instance of an Axes object in Matplotlib. It is used to specify the subplot or axes on which the scatter plot will be drawn. ax0 is assumed to be already defined in the code.

**scatter**
It is the variable that stores the returned scatter plot object. This object can be used for further customization or to add additional elements to the plot if needed.

**x=over_50['age']**
This specifies the data to be plotted on the x-axis of the scatter plot. Here, over_50['age'] is assumed to be a DataFrame or Series that contains the values to be plotted.

**y=over_50['chol']**
This specifies the data to be plotted on the y-axis of the scatter plot. Similarly, over_50['chol'] is assumed to be a DataFrame or Series that contains the values to be plotted.

**c=over_50['target']**
This parameter specifies the color of each individual data point in the scatter plot. Here, over_50['target'] is assumed to be a DataFrame or Series that provides the color values for each data point.

**cmap='winter'**
Thisis a parameter used to specify the colormap for the scatter plot.

A colormap, short for color map, is a mapping between numerical values and colors. It assigns colors to the data points based on their values, providing visual cues and aiding in the interpretation of the data.

The 'winter' colormap is one of the predefined colormaps available in Matplotlib. It ranges from blue to green, creating a cool color gradient. With this colormap, lower values will be represented by shades of blue, while higher values will be represented by shades of green.

By setting cmap='winter', the scatter plot will use the 'winter' colormap to assign colors to the data points based on the values in the 'target' column. This can enhance the visual representation of the data and make it easier to interpret.

You can explore and use different colormaps in Matplotlib depending on your specific visualization needs. **Some commonly used colormaps include 'viridis', 'jet', 'cool', 'hot', and many more**

In summary, the given code creates a scatter plot using the values from the "age" column as the x-axis, the values from the "chol" column as the y-axis, and the values from the "target" column as the color for each data point. The resulting scatter object can be further customized or used for additional plot manipulations if desired.

```python
#Customizing the graph style with seaborn
plt.style.use('seaborn-whitegrid')

# Creating the figure & the axes
fig, (ax0, ax1) = plt.subplots (nrows=2,
                                ncols=1,
                                figsize=(10,10))

#----------------- Add data to ax0-----------------------------------
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap = 'winter')
#Customize ax0
ax0.set(title='Heart Disease vs. Cholesterol Levels',
        xlabel="Age",
        ylabel="Cholesterol")

#Impose Limits on the X, y axis
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a mean line
ax0.axhline(over_50['chol'].mean(), linestyle ='--');



#------------------ Add the data & customize the ax1---------------
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap ='summer')
ax1.set(title="Heart Disease vs. Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

#Impose Limits on the X & y axis
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend for ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax1.axhline(over_50['thalach'].mean(), linestyle='--');

# -----------------Add a title to the figure------------------
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold')

```

### Important Note -axes limits vs. mean line (be careful)
The mean line may not appear on the graph in the given code snippet due to the potential mismatch in the y-axis limits.

In the code, the limits for the y-axis of ax0 are set using ax0.set_ylim([60, 200]). If the mean value of the 'chol' column, obtained from over_50['chol'].mean(), falls outside this range, the line may be present but not visible within the specified limits.

To ensure the mean line is visible, you can either adjust the y-axis limits to encompass the mean value or choose a different range for the limits. For example, you can remove the ax0.set_ylim([60, 200]) line or modify it to reflect a wider range if needed.
