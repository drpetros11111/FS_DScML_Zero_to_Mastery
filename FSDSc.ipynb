# 1st ploting method with pyplot API

**In Python's Matplotlib library, there are two primary ways to create plots: using the pyplot API and using the object-oriented (OO) interface. Here are the key differences between the two approaches**:

**Coding style**

The pyplot API follows a procedural coding style and is inspired by MATLAB. It provides a state-based interface where functions are called directly on the pyplot module. On the other hand, the OO interface is more object-oriented, where plot elements are represented as objects, and functions are called on those objects to modify them.

**Statefulness**

The pyplot API maintains an internal state, and each function call affects the global state of the current figure and axes. This makes it convenient for quick interactive plotting, but it can lead to confusion and potential issues when creating complex plots. In contrast, the OO interface explicitly defines and manipulates objects, making it more suitable for creating and customizing multiple figures and axes simultaneously without interference.

**Figure and Axes objects**

With the pyplot API, figures and axes are automatically created and managed behind the scenes. When you call functions like plt.plot(), the current figure and axes are inferred or created as needed. In the OO interface, you explicitly create and work with Figure and Axes objects. This provides more control and flexibility, especially when dealing with multiple subplots or custom layouts.

**Flexibility and customization**

The OO interface offers more flexibility for customization. You have direct access to all the plot elements as objects, allowing you to fine-tune properties and layouts. You can create and manipulate multiple subplots in a single figure, adjust axis scales, add annotations, and more. The pyplot API provides convenient shortcuts for common plotting tasks, but it may have limitations when it comes to advanced customization.

**In summary, the pyplot API is simpler and more convenient for quick and straightforward plots, while the OO interface offers greater control, customization, and scalability for complex plots and multi-figure layouts. The choice between the two approaches depends on your specific plotting needs and coding preferences.**


---
**Note**
The term "state-based interface"used above is not correct.** Matplotlib's pyplot API is not strictly state-based.**

Matplotlib's pyplot API is actually a collection of functions that provide a MATLAB-like interface for creating and customizing plots.** It does not maintain a global state that affects subsequent plot commands.** Instead, each plot command operates independently and does not modify the internal state of the library.

**However, pyplot does provide some automatic management of figures and axes behind the scenes.** For example, when you call plt.plot(), it will create a new figure and axes if they don't already exist. This behavior can give the impression of a stateful interface, as subsequent plot commands operate on the current axes or figure.

To clarify, the pyplot API does not have a persistent state that carries over between plot commands. Each function call is independent, and the library does not keep track of any state between commands

In regular expressions, certain characters and combinations have special meanings. Here are some commonly used characters and their meanings:

\d: Matches any digit (0-9).
\D: Matches any non-digit character.
\w: Matches any alphanumeric character (a-z, A-Z, 0-9, and underscore).

\s: Matches any whitespace character (spaces, tabs, line breaks).
\S: Matches any non-whitespace character.
.: Matches any character except a newline.
[]: Matches any single character within the brackets.
^: Matches the start of a string.
$: Matches the end of a string.
*: Matches zero or more occurrences of the preceding character or group.
+: Matches one or more occurrences of the preceding character or group.
?: Matches zero or one occurrence of the preceding character or group.
{n}: Matches exactly n occurrences of the preceding character or group.
{n,}: Matches n or more occurrences of the preceding character or group.
{n,m}: Matches between n and m occurrences of the preceding character or group.
These are just a few examples of the characters and syntax used in regular expressions. Regular expressions can become complex and powerful tools for pattern matching and text manipulation.

###Styling the graphs with seaborn's styles

The given code snippet is related to creating a scatter plot using the Matplotlib library. Let's break down the snippet:

**scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])**

**ax0:**
It represents an instance of an Axes object in Matplotlib. It is used to specify the subplot or axes on which the scatter plot will be drawn. ax0 is assumed to be already defined in the code.

**scatter**
It is the variable that stores the returned scatter plot object. This object can be used for further customization or to add additional elements to the plot if needed.

**x=over_50['age']**
This specifies the data to be plotted on the x-axis of the scatter plot. Here, over_50['age'] is assumed to be a DataFrame or Series that contains the values to be plotted.

**y=over_50['chol']**
This specifies the data to be plotted on the y-axis of the scatter plot. Similarly, over_50['chol'] is assumed to be a DataFrame or Series that contains the values to be plotted.

**c=over_50['target']**
This parameter specifies the color of each individual data point in the scatter plot. Here, over_50['target'] is assumed to be a DataFrame or Series that provides the color values for each data point.

**cmap='winter'**
Thisis a parameter used to specify the colormap for the scatter plot.

A colormap, short for color map, is a mapping between numerical values and colors. It assigns colors to the data points based on their values, providing visual cues and aiding in the interpretation of the data.

The 'winter' colormap is one of the predefined colormaps available in Matplotlib. It ranges from blue to green, creating a cool color gradient. With this colormap, lower values will be represented by shades of blue, while higher values will be represented by shades of green.

By setting cmap='winter', the scatter plot will use the 'winter' colormap to assign colors to the data points based on the values in the 'target' column. This can enhance the visual representation of the data and make it easier to interpret.

You can explore and use different colormaps in Matplotlib depending on your specific visualization needs. **Some commonly used colormaps include 'viridis', 'jet', 'cool', 'hot', and many more**

In summary, the given code creates a scatter plot using the values from the "age" column as the x-axis, the values from the "chol" column as the y-axis, and the values from the "target" column as the color for each data point. The resulting scatter object can be further customized or used for additional plot manipulations if desired.

```python
#Customizing the graph style with seaborn
plt.style.use('seaborn-whitegrid')

# Creating the figure & the axes
fig, (ax0, ax1) = plt.subplots (nrows=2,
                                ncols=1,
                                figsize=(10,10))

#----------------- Add data to ax0-----------------------------------
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap = 'winter')
#Customize ax0
ax0.set(title='Heart Disease vs. Cholesterol Levels',
        xlabel="Age",
        ylabel="Cholesterol")

#Impose Limits on the X, y axis
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

#Add a mean line
ax0.axhline(over_50['chol'].mean(), linestyle ='--');



#------------------ Add the data & customize the ax1---------------
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap ='summer')
ax1.set(title="Heart Disease vs. Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

#Impose Limits on the X & y axis
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend for ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax1.axhline(over_50['thalach'].mean(), linestyle='--');

# -----------------Add a title to the figure------------------
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold')

```
